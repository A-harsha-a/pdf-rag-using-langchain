```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("pdf")
documents = loader.load()

print(documents[1].page_content)
print(documents[1].metadata)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. Initialize the TextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, 
    chunk_overlap=200,
    separators=["\n\n", "\n", " ", ""]
)

# 2. Split the loaded documents
# He maps this to a custom function that processes lists of documents
chunks = text_splitter.split_documents(documents)
```

```python
for i in range (0,1):
    print(chunks[i])
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Model Loaded Successfully!")
```

```python
chunk_texts = [chunk.page_content for chunk in chunks]

vectors = embeddings.embed_documents(chunk_texts)

print("Number of embeddings:", len(vectors))
print("Dimension of each embedding:", len(vectors[0]))
```

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(chunks, embeddings)
```

```python
retriever = db.as_retriever(
    search_kwargs={"k": 3}
)
```

```python
query = input("Enter Query:")
```

```python
query
```

```python
docs = retriever.invoke(query)
```

```python
print(type(docs))
```

```python
for i, doc in enumerate(docs, 2):
    print(f"\nChunk {i}")
    print(doc.page_content)
```

```python
context = "\n\n".join(
    doc.page_content
    for doc in docs
)
```

```python
print(context)
```

```python
prompt = f"""
Answer the question using only the context below.

Context:
{context}

Question:
{query}
"""
```
